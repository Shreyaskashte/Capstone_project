## Extracting 'Close' column from each Stock

```python
data = [A_AAL, A_ALGT, A_ALK, A_DAL, A_HA, A_LUV,
        F_BCS, F_CS, F_DB, F_GS, F_MS, F_WFC,
        P_BHC, P_JNJ, P_MRK, P_PFE, P_RHHBY, P_UNH,
        T_AAPL, T_AMZN, T_FB, T_GOOG, T_IBM, T_MSFT,SNP500]

# Define the column names manually
column_names = ['A_AAL', 'A_ALGT', 'A_ALK', 'A_DAL', 'A_HA', 'A_LUV',
              'F_BCS', 'F_CS', 'F_DB', 'F_GS', 'F_MS', 'F_WFC',
              'P_BHC', 'P_JNJ', 'P_MRK', 'P_PFE', 'P_RHHBY', 'P_UNH',
              'T_AAPL', 'T_AMZN', 'T_FB', 'T_GOOG', 'T_IBM', 'T_MSFT','SNP500']

# Filter and extract the 'Close' column for each DataFrame and set 'Date' column as index.
filtered_data = []
for i, df in enumerate(data):
    df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
    df.set_index('Date', inplace=True)
# Get the data from past 5 Year's to analyse stocks.
    filtered_df = df.loc[df.index >= '2015-10-01', ['Close']]
    filtered_df.rename(columns={'Close': column_names[i]}, inplace=True)
    filtered_data.append(filtered_df)

# Merge the datasets based on the common column 'Close' using date index
merged_data = pd.concat(filtered_data, axis=1)

# Print the merged dataset
merged_data
```

## Calculating Beta Coefficients

```python
stock_return.columns
```

```python
returns = stock_return.dropna(axis=0)
```

```python
# Calculating Beta Coefficients for Stock Symbols with Respect to the S&P 500

betas = []
for stock_symbol in columns:
    stock_data = returns[stock_symbol]
    sp500_data = returns['SNP500']
        
    
    model = sm.OLS(stock_data, sm.add_constant(sp500_data)).fit()
    beta = model.params[1]  # Beta coefficient is the second parameter (index 1)
    betas.append(beta)
print(betas)
```

```python
selected_data['Beta coefficient'] = betas

# Print the updated DataFrame
selected_data
```

```python
# Exporting final stocks file to CSV:
selected_data.to_csv('data_stats1.csv',index=True)
```
